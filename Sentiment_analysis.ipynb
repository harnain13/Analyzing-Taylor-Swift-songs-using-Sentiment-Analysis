```python
valence_threshold = 0.5
energy_threshold = 0.5
```

```python
# lets create emotion labels:

def classify_emotion(row):
    if row['valence'] > valence_threshold and row['energy'] > energy_threshold:
        return 'Happy'
    elif row['valence'] < valence_threshold and row['energy'] < energy_threshold:
        return 'Sad'
    elif row['energy'] > energy_threshold:
        return 'Energetic'
    else:
        return 'Calm'
```

```python
taylor['emotion'] = taylor.apply(classify_emotion, axis=1)
print(taylor['emotion'].value_counts())
```

# Feature Engineering and Selection

```python
features = taylor[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 
                   'instrumentalness', 'liveness', 'valence', 'tempo']]
feature_names = taylor[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 
                   'instrumentalness', 'liveness', 'valence', 'tempo']].columns
labels = taylor['emotion']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

# Model Training and Evaluation

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, accuracy_score

X_train, X_test, y_train, y_test = train_test_split(features_scaled, labels, test_size=0.2, random_state=42)
```

```python
# a. Logistic Regression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
y_pred_log_reg = log_reg.predict(X_test)
print("Logistic Regression Report:")
print(classification_report(y_test, y_pred_log_reg))
print("Accuracy:", accuracy_score(y_test, y_pred_log_reg))
```

```python
# b. Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("Random Forest Report:")
print(classification_report(y_test, y_pred_rf))
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
# c. Neural Network
mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, early_stopping=True)
mlp.fit(X_train, y_train)
y_pred_mlp = mlp.predict(X_test)
print("Neural Network Report:")
print(classification_report(y_test, y_pred_mlp))
print("Accuracy:", accuracy_score(y_test, y_pred_mlp))
```

# improve model performance

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

cv_scores = cross_val_score(log_reg, features_scaled, labels, cv=5)
print("Cross-validation scores (Logistic Regression):", cv_scores)
print("Mean CV score:", cv_scores.mean())
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
grid_search_rf = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search_rf.fit(X_train, y_train)
print("Best parameters (Random Forest):", grid_search_rf.best_params_)
```

```python
y_pred_rf_tuned = grid_search_rf.predict(X_test)
print("Tuned Random Forest Report:")
print(classification_report(y_test, y_pred_rf_tuned))
print("Accuracy:", accuracy_score(y_test, y_pred_rf_tuned))
```

# Lets, Create a function that predicts the emotion of a song based on its features:

```python
best_rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=2
)

best_rf.fit(X_train, y_train)
```

```python
emotion_distribution = taylor['emotion'].value_counts()
print(emotion_distribution)
```

```python
def predict_emotion(song_features):
    """
    Predict the emotion of a song based on its features.
    
    Parameters:
    song_features (pd.DataFrame): A DataFrame containing the features of the song.
    
    Returns:
    str: The predicted emotion.
    """
    song_features_scaled = scaler.transform(song_features)
    predicted_emotion = best_rf.predict(song_features_scaled)
    return predicted_emotion[0]
```

```python
song_indices = [106, 101, 105, 72, 13]
```

```python
for song_index in song_indices:
    song_features = taylor.loc[[song_index], feature_names]
    song_name = taylor.loc[song_index, 'track_name']
    actual_emotion = taylor.loc[song_index, 'emotion']
    predicted_emotion = predict_emotion(song_features)
    print(f"Song Name: {song_name}")
    print(f"Actual Emotion: {actual_emotion}")
    print(f"Predicted Emotion: {predicted_emotion}\n")
```
